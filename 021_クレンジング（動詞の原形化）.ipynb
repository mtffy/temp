# 関数の定義

```python
nltk.download('wordnet')

# 動詞を原型に変換
def lemmatize_text(sr: pd.Series) -> pd.Series:
    w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
    lemmatizer = nltk.stem.WordNetLemmatizer()

    s_list = [lemmatizer.lemmatize(w, pos="v") for w in w_tokenizer.tokenize(sr)]
    sr_cleaned = ' '.join(s_list)
    return sr_cleaned
```

# 使用例

```python
# 使用方法
df_train['s_new_lem'] = df_train['s_new'].apply(str).apply(lemmatize_text)
df_train.head()
```
